```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv(r'D:\codebase\machine-learning-python\datasets\Advertising.csv')

df.head()
```

```python
X = df.drop('sales', axis=1)

X.head()
```

```python
y = df['sales']

y.head()
```

```python
polynomial_convertor_without_bias = PolynomialFeatures(degree=2, include_bias=False)

polynomial_convertor_with_bias = PolynomialFeatures(degree=2, include_bias=True)
```

```python
polyX_without_bias = polynomial_convertor_without_bias.fit_transform(X)

polyX_without_bias.shape
```

```python
polyX_with_bias = polynomial_convertor_with_bias.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(polyX_without_bias,y, test_size=0.3, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model_with_bias = LinearRegression(fit_intercept=True)
```

```python
linear_model.fit(X_train, y_train)
```

```python
pred_val = linear_model.predict(X_test)
```

```python
mean_absolute_error(y_test,pred_val)
```

```python
np.sqrt(mean_squared_error(y_test, pred_val))
```

```python
linear_model_with_bias.fit(X_train, y_train)
```

```python
pred_values_with_bias = linear_model_with_bias.predict(X_test)
```

```python
mean_absolute_error(y_test, pred_values_with_bias)
```

```python
np.sqrt(mean_squared_error(y_test, pred_values_with_bias))
```
