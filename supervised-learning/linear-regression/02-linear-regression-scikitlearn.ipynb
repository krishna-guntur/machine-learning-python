```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

from joblib import dump, load
```

```python
df = pd.read_csv(r'D:\codebase\machine-learning-python\datasets\Advertising.csv')

df.head()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=3, figsize=(20,10))

axes[0].plot(df['TV'], df['sales'],'o')
axes[0].set_xlabel('TV')
axes[0].set_ylabel('Sales')

axes[1].plot(df['radio'], df['sales'],'o')
axes[1].set_xlabel('radio')
axes[1].set_ylabel('Sales')

axes[2].plot(df['newspaper'], df['sales'],'o')
axes[2].set_xlabel('newspaper')
axes[2].set_ylabel('Sales')
```

```python
X = df.drop('sales', axis=1)

X.head()
```

```python
y = df['sales']

y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=101)
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, y_train)
```

```python
predicted_values = linear_model.predict(X_test)
```

```python
mean_absolute_error(y_test, predicted_values)
```

```python
np.sqrt(mean_squared_error(y_test, predicted_values))
```

You can save the model using Joblib dump, load methods

```python
final_model = linear_model.fit(X,y)
```

```python
dump(final_model, 'linear-regression.joblib')
```
